```python
#importing dependacies

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
#Load data sets

salary_data = pd.read_csv("salary_data.csv")

salary_data
```

```python
#plotting data

years = salary_data["YearsExperience"]

salary = salary_data["Salary"]

plt.plot(years, salary, marker='o', linestyle='-')

#  labels and title
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.title('Salary Growth by Experience')

# Display the graph
plt.grid(True)
plt.show()
```

```python
salary_data.info()
```

```python
# Define our x and y 

x = salary_data[["YearsExperience"]]
y = salary_data[["Salary"]]
```

```python
#Create model

model = LinearRegression()

#Then train a model

model.fit(x,y)
```

```python
#Predicting a model

predictions = model.predict(x)

print(predictions)
```

```python
# Predicting salaries for new employees  
new_experience = [[2], [4], [6], [8], [10]]

predicted_salaries = model.predict(new_experience)

print(predicted_salaries)
```

```python


# Scatter plot of actual data
plt.scatter(x, y, color='blue', label='Actual Data')

# Regression line
plt.plot(x, predictions, color='red', label='Regression Line')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction using Linear Regression")
plt.legend()
plt.show()
```

```python
# Evaluate the model 

from sklearn.metrics import r2_score

predictions = model.predict(x)

score = r2_score(y, predictions)

print("R² Score:", score)
```
